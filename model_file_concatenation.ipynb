```python
import os
import re
# import subprocess
```

```python
# directories

data_dir = "/Users/smb-uh/UHM_Ocean_BGC_Group Dropbox/Datasets/"

# model_run = 'cmip5'
model_run = 'cmip6_historical'
# model_run = 'cmip6_ssp585'

if model_run=='cmip5':
    cmip_dir = data_dir + "Model_Output/CMIP5/so/unconcatenated/"
    experiment = 'rcp85'
elif model_run=='cmip6_historical':
    cmip_dir = data_dir + "Model_Output/CMIP6/so/historical_download/unconcatenated/"
    experiment = 'historical'
elif model_run=='cmip6_ssp585':
    cmip_dir = data_dir + "Model_Output/CMIP6/so/ssp585_download/unconcatenated/"
    experiment = 'ssp585'

print(cmip_dir)
print(experiment)
```

```python
so_file_list = []

for file in os.listdir(cmip_dir):
    if file.endswith('.nc'):
        so_file_list.append(file)


# print(so_file_list)
# get a list of model names
unique_model_names = set()


# Regular expression to find the pattern
pattern = re.compile(r'Omon_(.*?)_'+experiment)

# Extract the text between "Omon_" and "_rcp85"
for file_name in so_file_list:
    match = pattern.search(file_name)
    if match:
        unique_model_names.add(match.group(1))

# Convert the set to a list
unique_values_list = list(unique_model_names)

# Print the unique values
print(unique_values_list)
```

```python
for model_name in unique_values_list:
    # now find file names in directory associated with each unique model name
    model_list = []

    for file in os.listdir(cmip_dir):
        if '_' + model_name + '_' in file:
            model_list.append(file)

    print(model_list)
    model_list_sorted = sorted(model_list)

    if len(model_list_sorted)>1:

        # find the start/end dates: 
        pattern = re.compile(r'_(\d{6})-(\d{6})\.nc')
        match = pattern.search(model_list_sorted[0])
        start_date = match.group(1)
        match = pattern.search(model_list_sorted[-1])
        end_date = match.group(2)


        # print(model_list_sorted[0][:-16])
        command = "ncrcat -h"

        for file in model_list_sorted:
            command = command + ' "' + cmip_dir+file + '"'

        out_filename = model_list_sorted[0][:-16] + start_date + '-' + end_date + '.nc'
        command = command + ' ' + '"' + cmip_dir +  out_filename + '"'
        print(command)
        os.system(command)

        # mv_command = "mv "
        mv_command = "mv " + '"' + cmip_dir + out_filename + '" ' + '"' + cmip_dir + '../' +  out_filename + '"'
        print(mv_command)
        os.system(mv_command)
```
